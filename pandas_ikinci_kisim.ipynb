```python
#Gruplama ve Özetleme (Groupby ve Aggregation)
import pandas as pd
df = pd.DataFrame({'kategori': ['A', 'B', 'A', 'B'], 'deger': [10, 20, 30, 40]})
gruplu = df.groupby('kategori')['deger'].sum()
print(gruplu)
```

```python
#Pivot Tablolar (Pivot Tables)
df = pd.DataFrame({'urun': ['A', 'B', 'A'], 'satis': [100, 200, 300], 'yil': [2020, 2020, 2021]})
pivot_tablosu = df.pivot_table(index='yil', values='satis', aggfunc='sum')
print(pivot_tablosu)
```

```python
#pandas ile Veri Görselleştirme
import matplotlib.pyplot as plt

df = pd.DataFrame({'urun': ['A','B','C'],'satis': [100,150,200]})
df.plot(kind = 'bar',x = 'urun',y = 'satis')
plt.show()
```

```python
#Zaman Serisi Verileri ile Çalışmak
df = pd.DataFrame({'tarih': ['2020-01-01','2020-02-01'], 'deger': [10,20]})
df['tarih'] = pd.to_datetime(df['tarih'])
df.set_index('tarih').plot()
plt.show()
```

```python
#pandas ile Veri Dönüştürme (Transformation)
df = pd.DataFrame({'sayi': [1,2,3,4]})
df['ikiyle_carpi'] = df['sayi'].apply(lambda x: x*2)
print(df['ikiyle_carpi'])
```

```python
#pandas ile Filtreleme ve Seçim
df = pd.DataFrame({'isim': ['Gamze','Nur','Aslan'], 'yas': [24,65,16]})
filtrelenmis_df = df[df['yas'] > 43]
print(filtrelenmis_df)
```

```python
#Kategorik Veriler ile Çalışmak
df = pd.DataFrame({'urun': ['A','B','A','C']})
df['urun'] = pd.Categorical(df['urun'])
print(df['urun'])
```
